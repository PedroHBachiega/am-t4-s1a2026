## Bibliotecas e Funções de Python

```python
import math

def equacao2grau(a,b,c):
  delta = b**2 - 4*a*c

  if (delta < 0):
    return 'Número sem Raíz', delta

  x1 = (-b + math.sqrt(delta)) / (2*a)
  x2 = (-b - math.sqrt(delta)) / (2*a)

  return x1, x2
```

## Testes de Função Matemática

```python
eq = equacao2grau(1, -3, 2)
print(eq)
```

```python
import random

for i in range(5):
  a = random.randint(-5, 5)
  b = random.randint(-10, 10)
  c = random.randint(-10, 10)

  eq += equacao2grau(a, b, c)
```

```python
for i in range(12):
  if(i % 2 == 0):
    if (eq[i] != 'Número sem raiz'):
      print('X1: [', eq[i],']')
  else:
    if (eq[i] != 'Número sem raiz'):
      print('X2: [', eq[i],']\n')
```
